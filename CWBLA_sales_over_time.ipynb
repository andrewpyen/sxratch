```python
storenvy = pd.read_csv(f'temp_data/{os.listdir("temp_data")[0]}', encoding='ISO-8859-1')
```

```python
storenvy['short_date'] = storenvy['Date'].str[:10].str.replace('/2023','')
```

```python
storenvy.short_date.value_counts()
```

```python
storenvy[storenvy['Product Name'].str.contains('Black Liberation')]['Product Quantity'].value_counts()
```

```python
CWBLA_sales = storenvy[storenvy['Product Name'].str.contains('Black Liberation')]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
CWBLA_sales_per_date = CWBLA_sales[['Order #', 'short_date', 'Product Quantity']].groupby('short_date').sum()
plt.figure(figsize=(50,20))
sns.set(font_scale=2.)
sns.lineplot(x='short_date', y='Product Quantity', data=CWBLA_sales_per_date)
```
