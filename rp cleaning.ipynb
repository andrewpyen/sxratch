```python
import pandas as pd
```

```python
csv = pd.read_csv('~/Downloads/Storenvy-the-rookery-press-All-Orders-February-19-2023.csv', encoding='ISO-8859-1')
```

```python
csv.info()
```

```python
csv['Name'] = csv['First Name'] + ' ' + csv['Last Name']
csv = csv.fillna(value={'Address 2':'Blank','Locale':'Blank','Shipping Status':'Payment Pending'})
csv['Address'] = csv['Address 1'] + ' ' + csv['Address 2'] + ' ' + csv['City'] + ', ' + csv['Locale'] + ' ' + csv['Postal Code'] + ' ' + csv['Country']
```

```python
# gpby = csv.groupby(by=['Order #','Product Name','Address','Shipping Status'], axis=0)[['Product Name', 'Product Quantity', 'Shipping Status']].sum()
```

```python
csv[['Order #','Product Name', 'Date', 'Address', 'Address 1', 'Name', 'Shipping Status', 'Product Quantity',]].head()
```

```python
product_counts = \
csv[['Order #','Product Name', 'Date', 'Address', 'Name', 'Shipping Status', 'Product Quantity',]]\
.set_index([csv.index, 'Order #', 'Date', 'Address', 'Name', 'Shipping Status','Product Name'])['Product Quantity'].unstack()
```

```python
only_product_counts = product_counts.reset_index().groupby(by=['Order #'])[csv['Product Name'].unique()].sum()
```

```python
join = pd.merge(csv[['Order #', 'Date','Address','Name','Shipping Status']], only_product_counts, how='left', on='Order #')
final = join.drop_duplicates()
```

```python
final.insert(loc=0,column='Order Source',value='Storenvy')
```
