# Linear Regression Steps

1. imports, load data
2. cleaning
3. EDA
4. prepare independent/dependent features
5. train/test split
6. scale feature vectors
7. model training
8. predition
9. performance metrics

```python
# import ml package
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline

# load in data
import pandas as pd

dataset = pd.read_csv('s3://yanxiang-gate/ml_practice_data/covid_worldwide.csv')
dataset.info()
```

```python
# data cleaning
```

```python
for c in dataset.columns:
    if c not in ['Serial Number','Country']:
        dataset[c] = dataset[c].str.replace(',','').astype(float)

dataset.info()
```

```python
nonnan = dataset.dropna()
```

```python
# data EDA
```

```python
nonnan.corr()
```

```python
import seaborn as sns

sns.pairplot(nonnan)
```

```python
# closer view of a single plot from the plot matrix above
plt.scatter(dataset['Total Deaths'], dataset['Total Recovered'])
```

```python
# single variable regression - NOT a prediction, only a visualization. 
# The ability to predict is what separates sklearns LinearRegression from sns.regplot
sns.regplot(x='Population', y='Total Deaths', data=nonnan)
```

```python
# 4. prepare independent / dependent features
```

```python
target_col = nonnan.pop('Total Deaths')

features_df = nonnan.drop(['Serial Number','Country'], axis=1)
# add target_col back to dataframe (in the last position of axis=1)
features_df.insert(len(features_df.columns), 'Total Deaths', target_col)
```

```python
features_df = features_df.drop(['Total Recovered', 'Total Test', 'Active Cases','Total Cases'], axis=1)

X = features_df.iloc[:,:-1]
y = features_df.iloc[:,-1]
```

```python
features_df
```

```python
# 5. train/test split
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
# 6. scaling features
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 7. Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
X.columns
```

```python
# check coefficients, intercept
reg.coef_, reg.intercept_
```

```python
# prediction
reg_pred = reg.predict(X_test)
```

```python
# scatter plot of predictions vs. test target

plt.scatter(reg_pred, y_test)
```

```python
# plotting residuals

residuals = y_test - reg_pred

sns.displot(residuals, kind='kde')
```

```python
# plotting predictions and residuals


# ideally the distribution is uniform (not clustered)
plt.scatter(reg_pred, residuals)
```

```python
# Performance metrics
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
reg.score(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
score
```
